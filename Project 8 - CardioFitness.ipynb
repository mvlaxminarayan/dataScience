```python
# univariate Analysis on Products sold
products = cardio_fitness_customers['Product']
plt.hist(products)
```

```python
# Univariate analysis on gender
gender = cardio_fitness_customers['Gender']
plt.hist(gender)
```

```python
#univariate analysis on Age
age = cardio_fitness_customers['Age']
sns.displot(age)
plt.axvline(age.mean(),color='b', linestyle='--') # get the mean line
plt.annotate('average = {}'.format(int(age.mean())),xy=(30,40)) # annotation
```

```python
# Univariate analysis on Marital status
marital_status = cardio_fitness_customers['MaritalStatus']
sns.displot(marital_status)
```

```python
# Univariate analysis on usage
usage = cardio_fitness_customers['Usage']
sns.displot(usage)
plt.axvline(usage.mean(),color='b', linestyle='--') # get the mean line
plt.annotate('average = {}'.format(round(usage.mean(),2)),xy=(3.5,60)) # annotation
```

Observation : Customers who wants to use the product 3 to 4 times are more.

```python
# Univariate analysis on fitness
fitness = cardio_fitness_customers['Fitness']
sns.displot(fitness)
plt.axvline(fitness.mean(),color='b', linestyle='--') # get the mean line
plt.annotate('average = {}'.format(round(fitness.mean(),2)),xy=(3.5,95)) # annotation
```

Observation : Customers with fitness 3 are more(~ 100) and customers with fitness 1 are very less( below 5).

```python
# Univariate analysis on miles
miles = cardio_fitness_customers['Miles']
sns.displot(miles)
plt.axvline(miles.mean(),color='b', linestyle='--') # get the mean line
plt.annotate('average = {}'.format(round(miles.mean(),2)),xy=(110,35)) # annotation
```

```python
def get_mean_value(x, **kwargs):
    # get mean value as annotation
    plt.annotate('avg={}'.format(round(x.mean(),2)),xy=(200,20),size=13)
g= sns.FacetGrid(cardio_fitness_customers,col='Product',height=4,aspect=.8,hue='Gender')
g.map(sns.scatterplot,'Miles','Age',alpha=.9)
g.map(get_mean_value,'Miles')
g.add_legend()
```

Observation : For the products(TM195 & TM498) most of the customers are under 35 age with average miles less than 88. And for the product(TM798) average miles is around 166

```python
def get_mean_value(x, **kwargs):
    # get mean value as annotation
    plt.annotate('avg Income= {}'.format(round(x.mean(),2)),xy=(60000,20))

g= sns.FacetGrid(cardio_fitness_customers,col='Product',height=4,aspect=.7,hue='Gender')
g.map(sns.scatterplot,'Income','Age',alpha=.7)
g.set_axis_labels('Income','Age')
g.add_legend()
g.map(get_mean_value,'Income')
#plt.annotate('avg Income= {}'.format(round(income.mean(),2)),xy=(60000,45)) # annotation
```

Observation : For the products(TM195 & TM498) most of the customers are under 35 age with average income less than 49000. And for the product the average income is around 75000 with huge difference in male & female

```python
#barplot for average fitness for each product
product_fitness = cardio_fitness_customers.groupby(by=['Product'])['Fitness'].mean().reset_index().sort_values(['Fitness'],
                    ascending=True) 
product_fitness.plot(x='Product',y='Fitness',kind='bar', figsize=(15,5))
plt.show()
```

observation : average fitness is high for TM798 product

```python
#barplot for average usage for each product
product_fitness = cardio_fitness_customers.groupby(by=['Product'])['Usage'].mean().reset_index().sort_values(['Usage'],
                    ascending=True)
product_fitness.plot(x='Product',y='Usage',kind='bar', figsize=(15,5))
plt.show()
```

observation : average usage is high for TM798 product

```python
#barplot for average miles for each product
product_fitness = cardio_fitness_customers.groupby(by=['Product'])['Miles'].mean().reset_index().sort_values(['Miles'],
                    ascending=True)
product_fitness.plot(x='Product',y='Miles',kind='bar', figsize=(15,5))
plt.show()
```

observation : average Miles is high for TM798 product
